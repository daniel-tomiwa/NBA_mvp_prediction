```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the mvp csv as a pandas dataframe
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "year", "Pts Won", "Pts Max", "Share"]]
mvps
```

```python
players = pd.read_csv("player_stats.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(50)
```

```python

players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head()
```

```python
players.groupby(["Player", "year"]).get_group(("Greg Anderson", 1991))
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"] #We pick out the row where the total stat is recorded
        row["Tm"] = df.iloc[-1, :]["Tm"]
        return row

players = players.groupby(["Player", "year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "year"])
```

```python
combined.head()
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined.head()
```

```python
teams = pd.read_csv("team_stats.csv")
```

```python
teams.head(20)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(20)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(10)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"]  = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "year"])
```

```python
stats.head()
```

```python
stats = stats.drop("Unnamed: 0", axis=1)
```

```python
stats.dtypes
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.isnull().sum()
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```

```python
highest_scoring.plot.bar("Player", "PTS")
```

```python
highest_scoring_year = stats.groupby("year").apply(lambda x: x.sort_values("PTS", ascending=False).head(1))
```
